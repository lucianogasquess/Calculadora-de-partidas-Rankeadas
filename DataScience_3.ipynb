```python
def producao_goiaba(regiao, pesos):
    result = 0
    for x, w in zip(regiao, pesos):
           result += x * w
    return result

  # Produção da região A
print(producao_goiaba(RegiaoA, pesos))
  # Produção da região B
print(producao_goiaba(RegiaoB, pesos))
```
